```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('nice_wine.csv')
df
```

```python
df.describe()
```

```python
X = df.drop('price', axis=1)
y = pd.DataFrame(df['price'])
```

```python
X.columns
```

```python
y.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

scaler = MinMaxScaler()
```

```python
scaled_X = pd.DataFrame(scaler.fit_transform(X), columns=['winery', 'wine', 'rating', 'region', 'body'])
scaled_y = pd.DataFrame(scaler.fit_transform(y))
scaled_y.rename(columns={0: 'price'}, inplace=True)
```

```python
scaled_X.head()
```

```python
scaled_y.head()
```

```python
scaled_X.describe()
```

```python
scaled_y.describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_X, scaled_y['price'], test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
```

```python
model.fit(X_train, y_train)

model.score(X_test, y_test)
```

```python
y_preds = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

print(f'Train Score:- {model.score(X_train, y_train)}')
print(f'Test Score:- {model.score(X_test, y_test)}')
print(f'MAE:- {mean_absolute_error(y_test, y_preds)}')
print(f'MSE:- {np.sqrt(mean_squared_error(y_test, y_preds))}')
```
